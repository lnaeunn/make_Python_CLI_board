```python
print("== CLI 게시판 실행 ==")

while True:
    cmd = input("명령어 ) ")

    if cmd == "exit":
      break

    elif cmd == "article write":
      input("제목 : ")
      input("내용 : ")
      print("1번 글이 생성되었습니다")
    else:
      print("지원하지 않는 기능입니다")

print("== CLI 게시판 종료 ==")
```
